```python
import pandas as pd

df=pd.read_csv('RecipeDataset/test_recipes.csv')
```

```python
print(df.columns)
```

```python
# Dataset shape and basic info
print("Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nFirst few rows:")
df.head()
```

```python
# Missing values analysis
print("Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0])
print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# Summary statistics for numeric columns
print("Summary Statistics:")
df.describe()
```

```python
# Unique values and value counts
print("Number of unique recipes:", df['Name'].nunique())
print("\nColumn value counts example (Servings):")
if 'Servings' in df.columns:
    print(df['Servings'].value_counts().head(10))
```

```python
# Split data into 80% and 20%
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Shuffle the dataframe
df_shuffled = df.sample(frac=1).reset_index(drop=True)

# Calculate split point
split_point = int(len(df_shuffled) * 0.8)

# Split into 80% and 20%
df_80 = df_shuffled[:split_point]
df_20 = df_shuffled[split_point:]

print(f"80% subset size: {len(df_80)} recipes")
print(f"20% subset size: {len(df_20)} recipes")

# Save both subsets to CSV files
df_80.to_csv('RecipeDataset/train_recipes_80.csv', index=False)
df_20.to_csv('RecipeDataset/test_recipes_20.csv', index=False)

print("\nFiles saved:")
print("✓ RecipeDataset/train_recipes_80.csv (80%)")
print("✓ RecipeDataset/test_recipes_20.csv (20%)")
```
